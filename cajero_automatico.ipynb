```python
# cajero automatico

class CuentaBancaria:

  def __init__(self, numero_cuenta, clave, saldo_inicial):
    self.numero_cuenta = numero_cuenta
    self.__clave = clave
    self.__saldo_inicial = saldo_inicial

  # valida clave
  def validar_clave(self, clave_ingresada):
    return clave_ingresada == self.__clave

  # verificar saldo
  def verificar_saldo(self):
    print(f"Saldo:  {self.__saldo_inicial}")

  # depositar
  def depositar(self, monto):
    if monto > 0:
      self.__saldo_inicial += monto
      print(f"Deposito exitoso. {monto}, Nuevo saldo: {self.__saldo_inicial}")
    else:
      print("Monto invalido")

  # retirar dinero
  def retirar(self, monto):
    if monto > self.__saldo_inicial:
      print("Saldo insuficiente")
    elif monto < 0:
      print("Monto invalido")
    elif monto > 0:
      self.__saldo_inicial -= monto
      print(f"Retiro exitoso. {monto}, Nuevo saldo: {self.__saldo_inicial}")
    else:
      print("Monto invalido")

  # cambiar clave
  def cambiar_clave(self, antigua_clave, nueva_clave):
    if self.validar_clave(antigua_clave) and len(nueva_clave) == 4 and nueva_clave.isdigit():
      self.__clave = nueva_clave
      print("Clave cambiada exitosamente")
    else:
      print("Error al cmabiar contraseña")


class ATM:

  def __init__(self):
    self.cuenta = {}

  def agregar_cuenta(self):
    numero_cuenta = input("Ingrese numero de cuenta: ")
    clave = input("Ingrese clave de 4 digitos: ")
    if len(clave) == 4 and clave.isdigit():
      saldo_inicial = float(input("Ingrese saldo inicial: "))
      self.cuenta[numero_cuenta] = CuentaBancaria(numero_cuenta, clave, saldo_inicial)
      print("Cuenta agregada exitosamente")
    else:
      print("Error al agregar cuenta")

  # autetiaccion de cuenta
  def autenticacion_cuenta(self):
    numero_cuenta = input("Ingrese numero de cuenta: ")
    clave = input("Ingrese clave: ")
    if numero_cuenta in self.cuenta and self.cuenta[numero_cuenta].validar_clave(clave):
      print("Autenticacion exitosa")
      self.menu_cuenta(numero_cuenta)
    else:
      print("Autenticacion fallida")
      return None

  def menu_cuenta(self, numero_cuenta):
    while True:
      print("\n Menu Cajero:")
      print("1. Verificar saldo")
      print("2. Depositar")
      print("3. Retirar")
      print("4. Cambiar clave")
      print("5. Salir")
      opcion = input("Ingrese una opcion: ")

      if opcion == "1":
        self.cuenta[numero_cuenta].verificar_saldo()
      elif opcion == "2":
        monto = float(input("Ingrese monto a depositar: "))
        self.cuenta[numero_cuenta].depositar(monto)
      elif opcion == "3":
        monto = float(input("Ingrese monto a retirar: "))
        self.cuenta[numero_cuenta].retirar(monto)
      elif opcion == "4":
        antigua_clave = input("Ingrese clave actual: ")
        nueva_clave = input("Ingrese nueva clave: ")
        self.cuenta[numero_cuenta].cambiar_clave(antigua_clave, nueva_clave)
      elif opcion == "5":
        print("Gracias por usar el cajero automatico")
        break
      else:
        print("Opcion invalida")


  def menu_principal(self):
    while True:
      print("\n Menu Principal:")
      print("1. Agregar cuenta")
      print("2. Autenticacion de cuenta")
      print("3. Salir")

      opcion = input("Ingrese una opcion (1-3): ")

      if opcion == "1":
        self.agregar_cuenta()
      elif opcion == "2":
        self.autenticacion_cuenta()
      elif opcion == "3":
        print("Gracias por usar el cajero automatico")
        break


# iniciar cajero automtico
if __name__ == "__main__":
  cajero = ATM()
  cajero.menu_principal()
```
